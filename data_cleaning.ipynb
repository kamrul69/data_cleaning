```python
# how to Replace the empty value
import pandas as pd
df = pd.read_csv('data.csv')

print(df.fillna(1000, inplace = True))
```

```python
# Replace only for specified column

import pandas as pd
df = pd.read_csv('data.csv')
df['Calories'].fillna(130, inplace = True)
```

```python
# Replace the mean, median or mode value for a specific column

import pandas as pd
df = pd.read_csv('data.csv')
x= df['Calories'].mean()
df['Calories'].fillna(x, inplace = True)

## median also like same way
```

```python
# for mode
import pandas as pd
df = pd.read_csv('data.csv')
x= df['Calories'].mode()[0]
df['Calories'].fillna(x, inplace = True)
```

### How to replacing data in a specific location

```python
df = pd.read_csv('/Users/mdkamrulhasan/Downloads/data.csv')
```

```python
print(df.dropna(inplace=True))
```

```python
print(df)
```

```python
## in location 69 the duration is 300. I want to replace 100 instead of 300
df.loc [69, 'Duration'] = 100
```

```python
print(df)
```

```python
# Replacing values by using loops
for x in df.index:
    if df.loc[x, 'Duration'] >120:
        df.loc[x, 'Duration'] = 100
```

```python
print(df)
```

```python
# removing rows by using loops

for x in df.index:
    if df.loc[x, 'Duration'] > 80:
        df.drop(x , inplace = True)
```

```python
print(df)
```
